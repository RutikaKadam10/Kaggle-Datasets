### Skewness 

Measures the asymmetry of the data distribution. It tells us whether the data is normally distributed or leaning towards the left or right.<br>

Skewness = 0 → Perfectly symmetric (normal distribution).<br>
Skewness > 0 → Right-skewed (Positive skew) (longer tail on the right).<br>
Skewness < 0 → Left-skewed (Negative skew) (longer tail on the left).<br>

```python
wine_quality_df.skew()
```

### Kurtosis

Kurtosis measures the tailedness of a distribution, indicating the presence of extreme values (outliers).<br>

Kurtosis = 3 → Mesokurtic (Normal distribution, moderate tails).<br>
Kurtosis > 3 → Leptokurtic (Heavy-tailed, many extreme values).<br>
Kurtosis < 3 → Platykurtic (Light-tailed, few or no extreme values).<br>

```python
wine_quality_df.kurtosis()
```

```python
#remove outliers using IQR Method
outlier_columns = ['volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','density','sulphates']

def remove_outliers(wine_quality_df, columns):
    for col in columns:
        Q1 = wine_quality_df[col].quantile(0.25)   #25th percentile
        Q3 = wine_quality_df[col].quantile(0.75)   #75th percentile
        IQR = Q3-Q1                                #Interquartile range
        
        #Define lower and upper bound
        lower_bound = Q1-1.5*IQR
        upper_bound = Q3+1.5*IQR
        
        #filer the dataframe
        wine_quality_df = wine_quality_df[(wine_quality_df[col]>=lower_bound)&(wine_quality_df[col]<=upper_bound)]
        
    return wine_quality_df 

#remove outliers from the selected columns
wine_quality_df = remove_outliers(wine_quality_df,outlier_columns)
```

```python
# Print the number of rows before and after outlier removal
print(f"Original Data Shape: {wine_quality_df.shape}")
print(f"Cleaned Data Shape: {wine_quality_df.shape}")
```

### Training model, Classification

```python
#classify function
from sklearn.model_selection import cross_val_score, train_test_split
def classify(model, data, label):
    x_train, x_test, y_train, y_test = train_test_split(data, label, test_size=0.25, random_state=42)
    #train model
    model.fit(x_train,y_train)
    print('Accuracy:',model.score(x_test,y_test)*100)
    
    #cross validation
    score = cross_val_score(model, data, label, cv=5)
    print('CV Score:',np.mean(score)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
classify(knn, data, label)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_reg = LogisticRegression()
classify(logistic_reg, data, label)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
classify(model, data, label)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
classify(model, data, label)
```
